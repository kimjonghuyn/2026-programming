```python
#산술연산자

print(7+4)
print(7-4)
print(7*4)#곱하기
print(7/4)#나누기
print(7**4)#거듭제곱
print(7//4)#나눗셈의 정수 몫
print(7%4)#나눗셈의 나머지
```

## 1. 산술 연산자 활용하기
**문제:** 사용자로부터 초 단위로 시간을 입력받아, 몇 분 몇 초인지 계산하여 출력하는 프로그램을 완성해 보세요.

```python
sec = int(input('시간을 초 단위로 입력하세요: '))

# 아래에 알맞은 산술 연산자를 사용하여 코드를 완성하세요.
minutes = sec//60 # 빈칸을 채우세요
seconds = sec%60 # 빈칸을 채우세요

print(f'{minutes}분 {seconds}초')
```

## 2. 비교 연산자와 논리 연산자
**문제:** 미리 저장된 아이디(`user123`)와 패스워드(`password456`)가 있습니다. 사용자로부터 아이디와 패스워드를 입력받아, 두 정보가 **모두** 일치하는지(`True` 또는 `False`) 확인하는 프로그램을 완성해 보세요.

```python
saved_id = 'user123'
saved_pw = 'password456'

userid = input('아이디를 입력하세요: ')
password = input('패스워드를 입력하세요: ')

# 비교 연산자(==)와 논리 연산자(and, or, not 중 하나)를 조합하여 완성하세요.
is_match = (userid == saved_id) and (password == saved_pw)
# 빈칸을 채우세요  (password == saved_pw)
print('id, pw 모두 일치 여부:', is_match)
```

## 3. 대입 연산자를 활용한 현금 자동 입출금기(ATM)
**문제:** 출금할 금액을 입력하면 50,000원, 10,000원, 5,000원, 1,000원 지폐가 최소 몇 장 필요한지 계산하는 프로그램입니다. 복합 대입 연산자(`%=`)를 활용하여 코드를 완성해 보세요.

```python
amount = int(input('출금할 금액을 입력하세요: '))

print(f'50,000원: {amount // 50000}장')
amount %= 50000  # 5만원권을 빼고 남은 금액을 다시 amount에 대입

print(f'10,000원: {amount // 10000}장')
# 알맞은 복합 대입 연산자를 작성하세요.
amount %= 10000

print(f'5,000원: {amount // 5000}장')
# 알맞은 복합 대입 연산자를 작성하세요.
amount %= 5000

print(f'1,000원: {amount // 1000}장')
amount %= 1000
```

## 4. 비트 연산자 확인
**문제:** 비트 시프트 연산자를 사용하여 정수 `4`에 대해 빠른 곱셈과 나눗셈을 수행해 보세요.

```python
value = 4

# 왼쪽으로 1비트 시프트 (곱하기 2의 효과)
result_left = value # 빈칸을 채우세요
print(f'{value} 왼쪽 1비트 시프트 결과: {result_left}')

# 오른쪽으로 1비트 시프트 (나누기 2의 효과)
result_right = value # 빈칸을 채우세요
print(f'{value} 오른쪽 1비트 시프트 결과: {result_right}')
```
