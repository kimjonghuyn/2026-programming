### [10쪽 예제 1] 자동으로 기사 만들기

```python
director = input('감독은 누구입니까? ')
stadium = input('경기장은 어디 인가요? ')
win = input('이긴 팀 이름은 무엇인가요? ')
loser = input('진 팀 이름은 무엇인가요? ')
score = input('스코어는 몇 대 몇 입니까? ')
vip = input('우승 선수는 누구입니까? ')

print('---- 주요뉴스 ------------')
print('오늘 %s에서 축구 경기가 열렸습니다. ' % stadium)
print('%s 감독이 이끄는 %s은 %s를 상대로 치열한 경기를 하였습니다. ' % (director, win, loser))
print('그 결과 %s(으)로 승리를 거두었고 우승 선수는 %s 선수입니다. ' % (score, vip))
```

### [12쪽 해 보기 1] 페트병 보상금 계산
- 금액을 담을 변수를 선언하고, 개수를 입력받아 정수로 변환한 뒤 총금액을 계산해 보세요.

```python
fee = 300  # 페트병 단가를 저장할 변수 선언
print(f'안녕하세요. 오늘의 투명색 페트병 금액은 {fee}원입니다.')
num = int(input('넣으실 투명색 페트병 개수를 입력해 주세요: '))  # 개수 변수 선언, 정수형으로 변환하여 입력받기
total = fee * num # 총금액 변수 선언 및 계산
print(f'넣으신 페트병 개수는 {num}개, 총금액은 {total}원입니다.')
print('아래에서 나오는 돈을 받아 주십시오.')
print('환경 보호에 동참해 주셔서 감사합니다.')
print('좋은 하루 보내세요.')
```
