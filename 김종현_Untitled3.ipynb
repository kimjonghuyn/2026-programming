#문제2
<font size=5>적금은 대표적인 금융 상품으로 일정 기간 금액을 납부한 후 이자와 함께 돌려받는 것을 말한다. 다음의 실행 결과를 보고 ‘적금 계산기’ 프로그램을 만들어 보자.
##입력예시

```
월 납입액을 입력하세요 (원): 100000
연 이자율을 입력하세요 (%): 3.75
적금 기간을 입력하세요 (월): 24
```

##출력예시
```
적금 기간 납입한 총 원금: 2400000 원
적금 기간 발생한 총 이자: 7500 원
적금 만기 시 받을 총금액: 2407500 원
```

````python
납입액 = int(input("월 납입액을 입력하세요"))
연이자율 = int(input("연 이자율을 입력하세요"))
적금기간 = int(input("적금 기간을 입력하세요"))
print(f"적금 기간 납입한 총 원금: {납입액}")
print(f"적금 기간 방생한 총 이자: {}")
print(f"적금 만기 시 받을 총금액: {납입액*(연이자율/100)*적금기간}")
````
