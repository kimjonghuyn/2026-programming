```python
#주민등록번호를 입력받은 후 태어난 연, 월, 일을 각각 풀력하는 프로그램을 완성해보자
jumin = input('주민등록번호 입력: ')
년 = "20"+jumin[:2]
월 = jumin[2:4]
일 = jumin[4:6]
print(f"연: {년}")
print(f"월: {월}")
print(f"일: {일}")
```

```python
#암호화
#replace => 문자 replace(old, new, [count])
#old = 현재 문자열에서 변경하고 싶은 문자
#new = 새로 바꿀 문자
#count = 변경할 횟수,
original = input("평문 입력: ")
encryption1 = original[::-1]
encryption2 = encryption1.replace('e','3').replace('o','0').replace('l','1')
print(encryption2)
```

```python
a = []
b = [10, 20, 30]
c = [1, 2, 'hello', 'hi']
d = [10, 2.5, ['hello', 'hi']]
print(d[2][0][0])
print(a,type(a))
print(b,type(b))
print(c,type(c))
print(d,type(d))
```

```python
#리스트 연산
a = [4, 2, 3]
b = [4, 5]
c = [1, 2, [3,4]]
print(a+b)
print(a*3)
print(a<b)
```

```python
#리스트 요소 변경
a = [1, 2, 3, 10, 20]
print('원본: ' , a)
a[2] = 100
print('결과:', a)
a[3:] = [4,5]
print('결과: ', a)
a[2:] = [10,11]
print('결과: ', a)
del a[2]
print('결과: ', a)
del a[1:]
print('결과: ', a)
del a[:]
print('결과: ', a)
```

```python
#range() -> range([start], stop, [step])
#start = 시작 값
#stop = 마지막 값 ==>  주의! 마지막 값 -1 stop = stop -1 까지 생성
#step = 증강값(간격)

list1 = list(range(10))
print(list1)
list2 = list(range(1,11))
print(list2)
list3 = list(range(1,11,2))
print(list3)
list4 = list(range(10,0,-2))
print(list4)
```
