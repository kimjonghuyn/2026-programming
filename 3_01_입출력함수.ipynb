#입력함수

input():키보드를 통해 값을 입력받을때 사용하는 함수

tip!

input()을 통해 입력된값은 문자열로 저장된다.

```python
a = input("이름: ")
b = input("나이: ")
print(a + "님 안녕하세요?")
print(20 - int(b) , "년 후면 20살이시네요!")
```

```python
a = int(input("가격: "))
b = int(input("할인율(%): "))
c = a * b / 100
print(f"할인받은 가격은 {c:0.0f}원입니다.")
```

```python
#섭씨( ℃)온도를 화씨 온도( ℉)로 변환하는 프로그램

a = float(input("섭씨 온도를 입력하세요:"))

b = (9 * a / 5) + 32

print(f"섭씨 {a}도는 화씨 {b}도 입니다.")
```
